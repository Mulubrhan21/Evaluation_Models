```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import numpy as np

# Step 1. Load dataset
def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv"))

housing = load_housing_data()

# Step 2. Separate features and target
housing_labels = housing["median_house_value"].copy()
housing = housing.drop("median_house_value", axis=1)

# Step 3. Define pipelines
numerical_cols = housing.drop("ocean_proximity", axis=1).columns.tolist()
categorical_cols = ["ocean_proximity"]

numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())
])

full_pipeline = ColumnTransformer([
    ('num', numerical_pipeline, numerical_cols),
    ('cat', categorical_pipeline, categorical_cols)
])

# Step 4. Train Linear Regression model
linear_regression_pipeline = Pipeline([
    ('preprocessor', full_pipeline),
    ('model', LinearRegression())
])

# --------- Evaluation Without Cross-Validation ---------
# Fit the model on the entire dataset
linear_regression_pipeline.fit(housing, housing_labels)

# Step 5. Evaluate the model on the same dataset (without cross-validation)
predictions = linear_regression_pipeline.predict(housing)
mse = mean_squared_error(housing_labels, predictions)
linear_rmse_no_cv = np.sqrt(mse)

# Print results for evaluation without cross-validation
print("Linear Regression RMSE (without cross-validation):", linear_rmse_no_cv)

# --------- Evaluation With Cross-Validation ---------
# Step 6. Evaluate the model using cross-validation
scores = cross_val_score(linear_regression_pipeline, housing, housing_labels, scoring="neg_mean_squared_error", cv=10)
linear_regression_rmse_scores = np.sqrt(-scores)

# Step 7. Print cross-validation results
print("Linear Regression RMSE scores (with cross-validation):", linear_regression_rmse_scores)
print("Mean RMSE (with cross-validation):", linear_regression_rmse_scores.mean())
print("Standard deviation of RMSE (with cross-validation):", linear_regression_rmse_scores.std())
```

```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
import numpy as np

# Step 1. Load dataset
def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv"))

housing = load_housing_data()

# Step 2. Separate features and target
housing_labels = housing["median_house_value"].copy()
housing = housing.drop("median_house_value", axis=1)

# Step 3. Define preprocessing pipelines
numerical_cols = housing.drop("ocean_proximity", axis=1).columns.tolist()
categorical_cols = ["ocean_proximity"]

# Full preprocessing pipeline
preprocessing = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), numerical_cols),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder())
    ]), categorical_cols)
])

# Step 4. Create the Decision Tree model pipeline
tree_reg_pipeline = make_pipeline(preprocessing, DecisionTreeRegressor(random_state=42))

# --------- Evaluation Without Cross-Validation ---------
# Step 5. Train the model on the entire dataset (without cross-validation)
tree_reg_pipeline.fit(housing, housing_labels)

# Step 6. Make predictions on the same dataset
predictions = tree_reg_pipeline.predict(housing)

# Step 7. Calculate RMSE without cross-validation
mse_no_cv = mean_squared_error(housing_labels, predictions)
DTree_rmse_no_cv = np.sqrt(mse_no_cv)

# Print results without cross-validation
print("Decision Tree RMSE (without cross-validation):", DTree_rmse_no_cv)

# --------- Evaluation With Cross-Validation ---------
# Step 8. Evaluate the model using 10-fold cross-validation
scores = cross_val_score(tree_reg_pipeline, housing, housing_labels, scoring="neg_mean_squared_error", cv=10)
decision_tree_rmse_scores = np.sqrt(-scores)

# Step 9. Print cross-validation results
print("Decision Tree RMSE scores (with cross-validation):", decision_tree_rmse_scores)
print("Mean RMSE (with cross-validation):", decision_tree_rmse_scores.mean())
print("Standard deviation of RMSE (with cross-validation):", decision_tree_rmse_scores.std())
```

```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split, cross_val_score
import numpy as np

# Step 1. Load dataset
def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv"))

housing = load_housing_data()

# Step 2. Separate features and target
housing_labels = housing["median_house_value"].copy()
housing = housing.drop("median_house_value", axis=1)

# Step 3. Define preprocessing pipelines
numerical_cols = housing.drop("ocean_proximity", axis=1).columns.tolist()
categorical_cols = ["ocean_proximity"]

preprocessing = ColumnTransformer([
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]), numerical_cols),
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder())
    ]), categorical_cols)
])

# --------- SVR Without Cross-Validation ---------
# Step 4. Create a SVR pipeline without cross-validation
svr_pipeline = make_pipeline(preprocessing, SVR())

# Step 5. Train the model on the entire dataset
svr_pipeline.fit(housing, housing_labels)

# Step 6. Make predictions on the same dataset
predictions = svr_pipeline.predict(housing)

# Step 7. Calculate RMSE without cross-validation
mse_no_cv = mean_squared_error(housing_labels, predictions)
svr_rmse_no_cv = np.sqrt(mse_no_cv)

# Print results without cross-validation
print("SVR RMSE (without cross-validation):", svr_rmse_no_cv)

# --------- SVR With Cross-Validation ---------
# Step 8. Evaluate the model using 5-fold cross-validation
svr_scores = cross_val_score(svr_pipeline, housing, housing_labels, scoring="neg_mean_squared_error", cv=5)
svr_rmse_scores = np.sqrt(-svr_scores)

# Step 9. Print cross-validation results
print("SVR RMSE scores (with cross-validation):", svr_rmse_scores)
print("Mean RMSE (with cross-validation):", svr_rmse_scores.mean())
print("Standard deviation of RMSE (with cross-validation):", svr_rmse_scores.std())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
import numpy as np

# Step 1. Load dataset
def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv"))

housing = load_housing_data()

# Step 2. Separate features and target
housing_labels = housing["median_house_value"].copy()
housing = housing.drop("median_house_value", axis=1)

# Step 3. Define pipelines
numerical_cols = housing.drop("ocean_proximity", axis=1).columns.tolist()
categorical_cols = ["ocean_proximity"]

numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())
])

full_pipeline = ColumnTransformer([
    ('num', numerical_pipeline, numerical_cols),
    ('cat', categorical_pipeline, categorical_cols)
])
# Step 4. Train svm model
train_set, test_set, train_labels, test_labels = train_test_split(housing, housing_labels, test_size=0.2, random_state=42)

# 3. Define the full pipeline with the SVR model (using default parameters)
svr_pipeline = Pipeline([
    ('preprocessor', full_pipeline),  # Preprocessing steps
    ('svr', SVR())  # Support Vector Regressor (default parameters)
])

# 4. Perform cross-validation
svr_scores = cross_val_score(svr_pipeline, train_set, train_labels, cv=5, scoring="neg_mean_squared_error")
svr_rmse_scores = np.sqrt(-svr_scores)

# 5. Print the results of cross-validation for SVR
print("SVR - Cross-validated RMSE scores:", svr_rmse_scores)
print("SVR - Mean RMSE:", svr_rmse_scores.mean())
print("SVR - Standard deviation of RMSE:", svr_rmse_scores.std())

# Step 6. Train the model on the entire training set
svr_pipeline.fit(train_set, train_labels)

# Step 7. Make predictions on the test set
svr_test_predictions = svr_pipeline.predict(test_set)

# Step 8. Evaluate the model on the test set (without cross-validation)
svr_test_rmse = mean_squared_error(test_labels, svr_test_predictions, squared=False)
svr_test_mae = mean_absolute_error(test_labels, svr_test_predictions)
svr_test_r2 = r2_score(test_labels, svr_test_predictions)

# Print results for test set evaluation
print("\nSVR Test set evaluation (without cross-validation):")
print("Test RMSE:", svr_test_rmse)
print("Test MAE:", svr_test_mae)
print("Test R² score:", svr_test_r2)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.pipeline import Pipeline
import numpy as np

# 1. Define the full pipeline with the SVR model
# 'full_pipeline' is your preprocessing pipeline (assuming it was defined earlier)
svr_pipeline = Pipeline([
    ('preprocessor', full_pipeline),  # Preprocessing steps
    ('svr', SVR())  # Support Vector Regressor
])

# 2. Manually specify hyperparameter values for tuning
C_values = [0.1, 1, 10, 100]  # Different values for regularization strength
kernel_types = ['linear', 'rbf']  # Different kernel types
gamma_values = ['scale', 'auto']  # Gamma for RBF kernel

# 3. Manual cross-validation and evaluation for each combination of hyperparameters
best_rmse = float('inf')
best_svr_model = None
best_params = {}

for kernel in kernel_types:
    for C in C_values:
        if kernel == 'rbf':
            for gamma in gamma_values:
                svr_pipeline.set_params(svr__kernel=kernel, svr__C=C, svr__gamma=gamma)
                scores = cross_val_score(svr_pipeline, train_set, train_labels, cv=5, scoring="neg_mean_squared_error")
                rmse_scores = np.sqrt(-scores)
                mean_rmse = rmse_scores.mean()
                
                print(f"Kernel: {kernel}, C: {C}, Gamma: {gamma}, RMSE: {mean_rmse} +/- {rmse_scores.std()}")
                
                if mean_rmse < best_rmse:
                    best_rmse = mean_rmse
                    best_svr_model = svr_pipeline
                    best_params = {'kernel': kernel, 'C': C, 'gamma': gamma}
        else:
            svr_pipeline.set_params(svr__kernel=kernel, svr__C=C)
            scores = cross_val_score(svr_pipeline, train_set, train_labels, cv=5, scoring="neg_mean_squared_error")
            rmse_scores = np.sqrt(-scores)
            mean_rmse = rmse_scores.mean()
            
            print(f"Kernel: {kernel}, C: {C}, RMSE: {mean_rmse} +/- {rmse_scores.std()}")
            
            if mean_rmse < best_rmse:
                best_rmse = mean_rmse
                best_svr_model = svr_pipeline
                best_params = {'kernel': kernel, 'C': C}

# 4. Output the best model and hyperparameters
print("\nBest SVR Model Hyperparameters:", best_params)
print("Best Cross-Validated RMSE:", best_rmse)

# 5. Train the best model on the entire training set
best_svr_model.fit(train_set, train_labels)

# 6. Evaluate the best model on the test set
test_predictions = best_svr_model.predict(test_set)

# Calculate RMSE, MAE, and R² score on the test set
test_rmse = mean_squared_error(test_labels, test_predictions, squared=False)
test_mae = mean_absolute_error(test_labels, test_predictions)
test_r2 = r2_score(test_labels, test_predictions)

print("\nTest set evaluation:")
print("Test RMSE:", test_rmse)
print("Test MAE:", test_mae)
print("Test R² score:", test_r2)
```
